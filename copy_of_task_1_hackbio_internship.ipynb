# Function 1: Translate DNA to Protein

Function 1: Translate DNA to Protein
Original DNA Sequence: "ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG"

Translation Process:

Codon Table Construction:

All combinations of triplets using T, C, A, G are mapped to their corresponding amino acids based on the standard genetic code.

Trimming the Sequence:

Ensures the DNA sequence length is a multiple of 3 (since codons are triplets).

Codon Translation:

Splits the DNA sequence into 3-letter codons and translates each using the codon table.

Unknown codons (not in the table) are marked with ?.

Resulting Protein Sequence: 'IRISIAR*TARR*'

```python

def translate_dna(dna_sequence):
    # Creating a codon table
    codon_keys = [a + b + c for a in "TCAG" for b in "TCAG" for c in "TCAG"]
    codon_values = list("FFLLSSSSYY**CC*WLLLLPPPPHHQQRRRRIIIIIIIIIMTTTTNNKKSSRRVVVVAAAADDEEGGGG")
    codon_table = dict(zip(codon_keys, codon_values))

    # Now ensuring the sequence length is a multiple of 3
    dna_sequence = dna_sequence[:len(dna_sequence) // 3 * 3]

    # Split sequence into codons and translate
    codons = [dna_sequence[i:i+3] for i in range(0, len(dna_sequence), 3)]
    protein_sequence = "".join(codon_table.get(codon, "?") for codon in codons)  # '?' if codon not found
    return protein_sequence

dna_seq = "ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG"
protein_seq = translate_dna(dna_seq)
protein_seq

```

#Function 2: Simulate Logistic Growth

Using three phases:

Lag Phase: Population stays roughly constant.

Exponential Phase: Population grows rapidly.

Stationary Phase: Growth slows as it approaches the carrying capacity
ùêæ
K.

Parameters:

K = 1000: Carrying capacity ‚Äî max sustainable population.

r = 0.5: Growth rate.

time_steps = 100: Number of time steps to simulate.

Randomized Phases:

lag_phase: Random between 5 and 20 steps.

exp_phase: Random between 20 and 50 steps.

After these, growth follows the logistic formula.

Output:

A dataframe with 100 simulations (Curve_ID 1 to 100), each with 100 time steps.

```python
import numpy as np
import pandas as pd


def simulate_logistic_growth(K=1000, r=0.5, time_steps=100):
    lag_phase = np.random.randint(5, 21)
    exp_phase = np.random.randint(20, 51)
    time = np.arange(1, time_steps + 1)
    pop = np.zeros(time_steps)
    pop[0] = 1

    for t in range(1, time_steps):
        if t < lag_phase:
            pop[t] = pop[t-1]
        elif t < lag_phase + exp_phase:
            pop[t] = pop[t-1] * np.exp(r)
        else:
            pop[t] = pop[t-1] + r * pop[t-1] * (1 - pop[t-1]/K)
        pop[t] = min(pop[t], K)

    return pd.DataFrame({'Time': time, 'Population': pop})

# Simulate 100 logistic growth curves
growth_curves = pd.concat([
    simulate_logistic_growth().assign(Curve_ID=i+1) for i in range(100)
], ignore_index=True)

growth_curves.head()
```

Function 3: Time to Reach 80% of Carrying Capacity


For each simulated growth curve, I found the first time step where the population reaches at least 80% of the carrying capacity (K = 1000), which is 800.

Steps:

Filter all time points where population ‚â• 800.

Group by Curve_ID.

Return the minimum time for each group ‚Äî i.e., the earliest point where the population hits the threshold.

Output: A summary table with two columns:

Curve_ID: ID of the simulated growth curve.

Time: Time step when population first reached 80% of
ùêæ
K.



```python

def time_to_80_percent_K(df, K=1000):
    threshold = 0.8 * K
    df_filtered = df[df['Population'] >= threshold]
    return df_filtered.groupby('Curve_ID', as_index=False)['Time'].min()

# Apply the function to the growth curves
time_80_percent = time_to_80_percent_K(growth_curves)
time_80_percent.head()
```

Hamming distance between slack and x user name

```python

def hamming_distance(str1, str2):
    """Calculates the Hamming distance between two strings."""
    distance = 0
    for i in range(min(len(str1), len(str2))):
        if str1[i] != str2[i]:
            distance += 1
    return distance

slack_name = "Victor"
twitter_handle = "Savick"

# Calculate distance
distance = hamming_distance(slack_name, twitter_handle)
distance
```
